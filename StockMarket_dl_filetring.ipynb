```python

import re
import tweepy
import pandas as pd
from tweepy import OAuthHandler
from textblob import TextBlob
import numpy as np
import matplotlib
import configparser
```

```python
config= pd.read_csv("/config.csv")
```

```python
twitterApiKeys= config['twitterApiKey'][0]
twitterApiSecret=config["twitterApisecret"][0]
twitterApiAccess=config['twitterApiAccessToken'][0]
twitterApiAccessTokenSecret=config['twitterApiAccessTokenSecret'][0]
```

```python
auth=tweepy.OAuthHandler(twitterApiKeys,twitterApiSecret)
auth.set_access_token(twitterApiAccess,twitterApiAccessTokenSecret)
twitterApi=tweepy.API(auth,wait_on_rate_limit=True)
```

```python
keywords='#Omicron'
limit=2000
tweets=tweepy.Cursor(twitterApi.search,q=keywords,count=100,tweet_mode='extended').items(limit)
#creating dataframe
columns=['User','Tweet']
data=[]

# tweets=twitterApi.user_timeline(screen_name=user,count=limit,tweet_mode="extended").items(limit)

for tweet in tweets:
  # print(tweet.full_text)
  data.append([tweet.user.screen_name,tweet.full_text])


df=pd.DataFrame(data,columns=columns)
print(df.head())
```

```python
def cleanup(text):
    temp = text.lower()
    temp = re.sub("'", "", temp) # to avoid removing contractions in english
    temp = re.sub("@[A-Za-z0-9_]+","", temp)
    temp = re.sub("#[A-Za-z0-9_]+","", temp)
    temp = re.sub(r'http\S+', '', temp)
    temp = re.sub('[()!?]', ' ', temp)
    temp = re.sub('\[.*?\]',' ', temp)
    temp = re.sub("[^a-z0-9]"," ", temp)
    temp=re.sub(r'RT : ','',temp)
    return temp
```

```python
df['Tweet']=df['Tweet'].apply(cleanup)
```

```python
def getTextSubjectivity(text):
  return TextBlob(text).sentiment.subjectivity
```

```python
def getTextPolarity(text):
  return TextBlob(text).sentiment.polarity
```

```python
df['subjectivity']=df['Tweet'].apply(getTextSubjectivity)
df['polarity']=df['Tweet'].apply(getTextPolarity)
```
